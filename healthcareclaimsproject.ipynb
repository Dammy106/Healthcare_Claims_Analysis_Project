```python
### Feature Engineering
```

```python
# Creating age group bins
bins = [0, 30, 50, 100]
labels = ['Young Adult', 'Middle-Aged', 'Senior']
df['agegroup'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
# BMI group classification (based on standard BMI categories)
bmi_bins = [0, 18.5, 24.9, 29.9, 100]  # thresholds for underweight, normal, overweight, and obese
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
df['bmigroup'] = pd.cut(df['bmi'], bins=bmi_bins, labels=bmi_labels)
```

```python
# Log-transforming claim amounts to normalize the distribution
df['Log_Claims'] = np.log(df['Claim'] + 1)
```

```python
## Now time to encode data (Transform categorical to Numerical)
```

```python
#Want to convert all categorical to numerical data
#Making sure all values are known, especially for the region
print(df['gender'].unique())
print(df['diabetic'].unique())
print(df['smoker'].unique())
```

```python
#Now convert
df['gender'] = df['gender'].map({'male': 1, 'female': 0})
df['diabetic'] = df['diabetic'].map({'Yes': 1, 'No': 0})
df['smoker'] = df['smoker'].map({'Yes': 1, 'No': 0})
```

```python
# Apply one-hot encoding without dropping any regions
df = pd.get_dummies(df, columns=['region'], drop_first=False)

# Convert boolean columns (region-related) to 1/0
region_columns = [col for col in df.columns if 'region' in col]
df[region_columns] = df[region_columns].astype(int)
```

```python
df.info()
```

```python
# One-hot encoding Age_group and BMI_group
df = pd.get_dummies(df, columns=['agegroup', 'bmigroup'], drop_first=True)

df.head()
```

```python
# Converting boolean columns to integers
for col in df.columns:
    if 'agegroup' in col or 'bmigroup' in col:
        df[col] = df[col].astype(int)
df.info()
```

```python
# Log transforming the Claim column to reduce skewness
df['log_claim'] = np.log(df['claim'] + 1)  # Add 1 to avoid log(0)

# Checking the result
df[['claim', 'log_claim']].head()
```

```python
df.info()
```

```python
# Dropping the 'claim' column. This way is might be easier actually
df = df.drop(columns=['claim'])
```

```python
df = df.drop(['age','bmi'], axis=1)
```

```python
numerical_columns = [ 'bloodpressure', 'children', 'log_claim']
```

```python
### Done with prepocessing, Feature Engineering etc. time for clustering
```

```python
from sklearn.cluster import KMeans


wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Elbow method found 4 as optimal clusters
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(df)

#added clusters to df
df['Cluster'] = kmeans.labels_
```

```python
# Grouping by clusters and calculating mean for each feature
cluster_summary = df.groupby('Cluster').mean()
print(cluster_summary)
```

```python
#General Observations:
#Smoking and Obesity are key drivers of high healthcare claims, especially in Cluster 3.
#Family Size (Number of Children) does not appear to be a major determinant of higher claims, as seen in Cluster 2.
#Geography: The Southeast region has a large concentration of low-claim individuals (Cluster 1), while Cluster 3, with high claims, is spread across multiple regions.
```

```python
# Bar plot for average log-claim by cluster
cluster_avg_claim = df.groupby('Cluster')['log_claim'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Cluster', y='log_claim', data=cluster_avg_claim)
plt.title('Average Log-Claim Amount by Cluster')
plt.show()
```
